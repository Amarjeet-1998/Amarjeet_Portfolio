```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_excel(r"/content/CDS View.XLSX")
df1.head()
```

```python
df2 = pd.read_excel(r"/content/Project_information.xlsx")
df2
```

```python
df1.shape
```

```python
df1.columns
```

```python
df1.dtypes
```

```python
df1['Status'] =np.where(df1['ZUNITCOUNT']== 1,"Success","Cancel")
df1.head()
```

```python
df1["Cancel Unit"] =np.where(df1['Status'] =="Cancel",df1['ZUNITCOUNT'],0)
df1["Success Unit"] = np.where(df1['Status']=="Success",df1['ZUNITCOUNT'],0)
df1["Cancel Builtup Area"]=np.where(df1['Status']=="Cancel",df1["Superbuiltup area"],0)
df1["Success Builtup Area"]=np.where(df1['Status']=="Success",df1["Superbuiltup area"],0)
df1["Cancel Sale values"] =np.where(df1['Status']=="Cancel",df1["Sale Value"],0)
df1["Success Sale values"] =np.where(df1['Status']=="Success",df1["Sale Value"],0)
```

```python
df1.head(20)
```

```python
df2.columns
```

```python
df1.columns
```

```python
Combine_files = pd.merge(df2,df1,left_on='Profit Center',right_on='Plant')
Combine_files.head(10)
```

```python
print(Combine_files.dtypes)
```

```python
Combine_files['Year'] =pd.DatetimeIndex(Combine_files['REPORT_MONTH']).year
```

```python
Combine_files['Quarter'] =pd.to_datetime(Combine_files['REPORT_MONTH']).dt.quarter
```

```python
Combine_files['Month'] =pd.to_datetime(Combine_files['REPORT_MONTH']).dt.month
```

```python
Combine_files.head(10)
```

```python
Combine_files['Net Unit'] = Combine_files["Success Unit"]+Combine_files['Cancel Unit']
Combine_files['Net Builtup Area'] = Combine_files['Success Builtup Area']+Combine_files['Cancel Builtup Area']
Combine_files['Net Sale Amount'] = Combine_files['Success Sale values']+Combine_files['Cancel Sale values']
```

```python
Combine_files['Avg Rt/Sqft'] = Combine_files['Net Sale Amount']/Combine_files['Net Builtup Area']
```

```python
Pivot = pd.pivot_table(Combine_files,values=['Cancel Unit','Cancel Builtup Area','Cancel Sale values','Success Unit','Success Builtup Area','Success Sale values','Net Unit','Net Builtup Area','Net Sale Amount','Avg Rt/Sqft'],
                       index=['Type of Project','PROJECT STATUS','Description'],aggfunc=[np.sum],fill_value=0)
```
